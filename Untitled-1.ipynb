```r

process_dataset <- function(df, pattern) {
  df <- df %>%
    select(text) %>%
    mutate(negative = as.integer(grepl(pattern, text, ignore.case = TRUE)))
  return(df)
}

# Apply the function to each dataset in the list
processed_datasets <- lapply(list, process_dataset, pattern)
```
